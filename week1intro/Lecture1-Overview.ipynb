## Some data

Here is a kind of graph which often gets trotted out in debates about questions like this. 

Rather than just show the graph, I'm going to present some code which will produce it. If you have already taken Data 8 most of this should look familiar. If not, don't sweat it, we'll be walking you through how to making graphs like this later in the course.

```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing country-level gun data
gundata = pd.read_csv('data/guncountries.csv')
# Making scatterplot of gun ownership/gun deaths
sns.scatterplot(x='per100', y='total', data=gundata)
# Labelling the axes
plt.title("Gun Ownership and Gun Deaths, OECD Countries")
plt.xlabel("Gun Ownership per 100 people (2019)")
plt.ylabel("Gun Homicides per 100,000 people (2021)")
```

We’ll learn more about the details of how to make and interpret graphs like this later on, but for now let’s focus on the basics. This is called a “scatterplot”, and it contains a bunch of dots which all correspond to countries. The “Guns Ownership per 100 people” near the bottom tells us that this is what the horizontal axis (or x axis) measures, meaning points further to the right correspond to countries with higher gun ownership. The label on the left which says “Gun Homicides deaths per 100,000 people” tells us that this is the measure on the vertical axis (or y axis), meaning higher points correspond to countries with more gun related deaths. 

Why might a graph like this be informative about gun policy? A pretty clear pattern arises that countries further to the right tend to be higher. We call this a positive correlation (or a positive relationship). In this case, the pattern tells us that countries with more guns tend to have more gun-related deaths. This is why gun-control advocates like showing pictures like this.

Further, gun control advocates in the United States are particularly apt to show graphs like this, because the point way off in the top right corresponds to the U.S That is, the US has an unusually high degree of gun ownership and gun homicides. So maybe trying to reduce gun ownership would make us safer.

Even if you think this conclusion is correct, you can hopefully see some reasons why we may not be ready to  jump to it. Maybe the United States is just different for other reasons, and we can’t learn a ton by comparing across countries that are very different on many dimensions. 

Ok then, let’s look just within the United States. Here is a graph that plots states using some similar measures. The horizontal axis is now what proportion of households that have at least one gun, which is a bit different than guns per person, but still the general idea is that states to the right tend to have higher gun ownership, and again states which are higher vertically have more gun-related deaths. 


```python
# Loading state level data
gunstate=pd.read_csv("data/gunstate2016.csv")
# Making scatterplot
sns.scatterplot(x='HFR', y="Deathrate", data=gunstate)
# Labeling axes
plt.title("Gun Ownership and Gun Deaths by US State, 2016")
plt.xlabel("Proportion of Households with a Gun")
plt.ylabel("Gun Homicides per 100k people")
```

Again, we see a positive correlation: states with more guns have more gun-related deaths. However, just like with the last picture, this doesn’t prove that it is the higher prevalence of guns that is causing more gun-related deaths. A quick way to summarize why, which I hope you have heard before is that correlation does not imply causation. A lot of what we will do in this course is think carefully about what this phrase means, why it is true, and what we can do about it.

A related theme that we’ll explore quite a bit is that if we want to learn useful things from data -- for example, what causes what -- we need to think theoretically. So let’s think of some theories that might explain this data.

## The Importance of Theory

A first theory, which we’ll call the “anti-gun” theory, goes as follows. No matter how many guns are in a society, people are going to get in arguments, commit crimes, and otherwise be in scenarios that might turn violent. If one or more people involved in these scenarios possess a gun, an interaction that might otherwise just result in yelling or maybe a fist fight could escalate to a shooting. More generally, having more guns out there just leads to more opportunities for them to be used to kill someone. So, this theory predicts that if we were to remove some guns from our society, there would be fewer homicides in general, and gun-related homicides in particular.

Here is a second theory, which we’ll call the “pro-gun” theory. Regardless of whether or not you like it in the end, this theory has a starting point which I’ll try to convince you is a very powerful one: let’s first ask why people *choose* to own guns. In the “anti-gun” theory we talked about guns being injected or removed into a society as if by magic, but the level of gun ownership in a society is ultimately driven by choices made by humans who have reasons to own a gun or not. 

A Pew poll in 2017 asked gun owners in the U.S. what reasons they had for owning a gun (they could give multiple reasons). Here is a *bar plot* of their responses. Each bar tells us how many gun owners per 100 gave each reason.

```python
# Loading gun survey data
gunreasons = pd.read_csv("data/gunreasons.csv")
# Plotting reasons given for gun ownership
sns.barplot(x="Proportion", y="Reason", data=gunreasons)
```

As we can see the most common reason by a decent margin is "protection". Like with any survey, there might be reasons not to take this at face value, but let’s take people at their word that many own guns for protection. Where will people be most motivated to buy guns for protection? There are certainly many factors, but a big one is how dangerous one’s neighborhood, county, or state is. So people who live in areas with lots of violence -- including gun-related homicides -- may have more reason to own guns. In other words, the causation could flow in the other direction: living in a dangerous place leads people to buy guns. Pro-gun activists often take this argument a step further and say that if lots of people own guns, one may be more reluctant to get into a heated argument or rob a house knowing the owner might have a handgun under their pillow. That is, high gun ownership might deter violent situations to start with, which could weaken but not eliminate the positive correlation we observe. 

Regardless of whether you find either of these theories convincing, they could in principle explain the patterns we observed in the two figures. And we can certainly come up with more. A “neutral” theory might be that some places just have a more violent culture than others which leads to both more gun ownership and more gun-related homicides, even if neither of those two variables directly causes the other to increase or decrease. 

Why is it important to think through these different theories? One reason is pure curiosity: as social scientists, we want to understand how the world works. That is, we are not particularly interested in how dots line up in graphs; we want to answer questions like why some places are more violent than others, and why some places have higher gun ownership than others. And the different theories we discussed above -- which again could all produce the same dots on the graph -- reflect very different answers to these questions.



## Theory back to data 

A more practical reason we need to think theoretically is that we often debate questions like this when asking policy questions, like should we pass laws that make it harder to buy guns, or increase the penalties associated with illegal possession. If the “pro-gun” theory is true, then doing so will reduce homicides. If the “anti-gun” theory is true, then doing so will be useless if not counterproductive. (Also, setting aside any societal decisions, which theory is true also has implications for whether or not you should choose to own a gun!)

So what can we do to figure out which theory is right, or what is causing what? There are lots of possible approaches, and we’ll spend much of the course discussing these. 

One possibility is to try and find other data which can more directly tell us what causes what. To see how, let’s think a bit about the core problem here, which we will discuss more in week 6. The problem is that places with more and fewer guns may be different on other dimensions too. Some are more violent than others, and some have factors that make individuals more likely to buy guns than others. So we can’t just attribute the difference in gun homicides to the difference in gun ownership, since it might be driven by these other factors.

A broad strategy this suggests -- and we will see this play out in various ways -- is to try and find some places which are very similar to each other, other than in gun ownership. Intuitively, if we could find pairs of towns that are otherwise identical except among each pair one has much higher gun ownership than the other, we can be more comfortable attributing differences in homicides to this fact. 

The most powerful tool we have as social scientists to get comparisons like this -- that is, having two groups who are generally similar other than the factor we want to study -- is to run an experiment. Hypothetically, if we took two groups of towns that are otherwise similar on average, but then gave a bunch of guns away in one group but not the other (perhaps flipping a coin to decide which towns get the guns), by comparing their homicide rates afterwards we can be pretty sure that it was our intervention that caused any differences. 

However, there is a reason why I framed this as a hypothetical: it is certainly unethical and also probably infeasible to randomly add or subtract guns from different areas. And this is true for many of the questions we study in political science, like whether more democratic countries are more likely to fight wars, or whether nonviolent political movements are more likely to achieve their goals. 

Even though we can’t literally run experiments, we can draw inspiration from what makes experiments work, and try to find different areas we can compare that we know are pretty similar other than in gun ownership.

## A Natural Experiment?

One variant of this which is relatively easy if not perfect is to compare the same place before and after they make a policy change. For example, after a <a href="https://en.wikipedia.org/wiki/Port_Arthur_massacre_(Australia)">mass shooting in 1996</a>, Australia passed sweeping laws that, among other things, banned certain automatic and semi-automatic weapons. Let's see what happened to gun deaths after the shooting/passing of the law:

```python
# Loading up the data
aus = pd.read_csv("data/aus_guns.csv")
# Subsetting to years after 1988
aus = aus[aus['Year'] > 1988]
# Plotting trend in gun deaths
sns.lineplot(x='Year', y='Deathrate', data=aus)
# Adding annotations
plt.axvline(1996, linestyle="--")
plt.title("Trends in Gun Deaths per 100k people in Australia")
```

The vertical dotted line corresponds to the year the shooting happened and the law was passed. After that, the number of gun deaths fell. 

Recall the ideal comparison would involve similar places, which only differ in gun laws. Does this count? Certainly “Australia before 1996” and “Australia after 1996” are pretty comparable in a lot of ways: they are both Australia! So we might be able to learn something from these data.

However, there are some problems. First, Australia passed a series of gun control regulations in 1996. So it is pretty hard if not impossible to attribute the decrease in gun deaths to any particular one of these reforms. 

To see that this is a common issue, here is a very different story. My wife and I have a wonderful dog, named Ladybear. Ladybear is getting pretty old, and like a lot of old dogs she has some back problems. A couple years ago these flared up and she had a hard time walking. When you have an amazing dog who is in pain, it’s hard to not do everything you can to fix the problem. So we did a lot of things. We got her supplements, dog ibuprofen, and took her on shorter walks. And since this is the Bay Area, we gave her CBD and took her to acupuncture. (In case you are worried about learning about research methods from someone who believes in giving dogs CBD acupuncture,  there is some <a href="https://www.ncbi.nlm.nih.gov/pmc/articles/PMC5556488/">reasonable evidence</a> that these work!) After throwing all these solutions at the problem, she got much better, and still is, which is a relief. But the problem is we don’t know what solved the problem, which is bad for us, but good for the hippie dog treatment industry. 

Back to guns in Australia, another reason we should be hesitant to conclude that even the suite of policies passed caused the decrease in gun violence is that other things changed as well, beyond the policies. For example, people may have been more careful and wary of guns after the traumatic event. Third, by just comparing one country over two time periods, there could have been other random factors that led to more or fewer gun deaths in these periods. 

## A Better Natural Experiment

Overall, there are certainly some things we can learn from the Australia case, but we can’t reach strong conclusions about whether the availability of guns leads to more homicides. The limitations of this analysis also hints at a way we can do better: by trying to find a case where the ease of gun ownership changed in one area for reasons outside of their control.

A <a href="https://www.cambridge.org/core/journals/american-political-science-review/article/abs/crossborder-spillover-us-gun-laws-and-violence-in-mexico/438E607A07F32D57AF244B61ED38FB28">paper by Dube, Dube, and Garcia-Ponce</a> does just this. Here is the case they study. In 2004, a Federal Assault Weapon Ban expired in the US, meaning that certain kinds of guns that were previously not sold to civilians became available. However, some states passed laws which kept the ban in place within their borders. So, in effect, there was no policy change in some states like California, while the guns became easier to buy in other states like Texas, New Mexico, and Arizona.

Why do I mention these four states? Because they are the four states which border Mexico, and guns from the US were frequently smuggled across the border and used in the ongoing drug war there. In effect, we get something like the unethical experiment mentioned above where a bunch of guns become more available in some places but not others. (But since the researchers weren't the ones changing the gun policy, minus the ethical concerns.)

A key caveat is that parts of Mexico close to California might be different in other ways. This is not just a hypothetical; the areas of Mexico close to the Californian border were particularly violent at this time. However, we can combine this with the idea from the Australia example, and compare how homicide rates changed in parts of Mexico close to California vs part of Mexico close to the other states. Since the change in gun availability in these two places was unrelated to anything in Mexico, it is reasonable to assume that changes in homicides in different parts of Mexico were primarily driven by the change in assault weapon availability. 

Dube, Dube, and Garcia-Ponce do just this. Here is a replication of one of their main graphs.

```python
df = pd.read_csv('data/nca_deaths.csv')
# Plotting gun-related homicides per year for municipios adjacent to CA
sns.lineplot(x='year', y='homdguns', data=df[df['nonca']==0], label='Adjacent to CA')
# Plotting homicides per year for municipios not adjacent to CA
sns.lineplot(x='year', y='homdguns', data=df[df['nonca']==1], label='Adjacent to TX, AZ, or NM')
#Setting the y axis to go from 0 to 700
plt.ylim(0,700)
# Adding a title to the plot
plt.title("Gun-related Homicides")
# Adding a legend to the plot
plt.legend()
# Making a vertical line when the ban expired
plt.axvline(2004, color='black', linestyle='--')
```

This graph plots the sum of gun related homicides in the parts of Mexico close to California with a blue line, and those close to the other states in a orangle line. The number of homicides in cities close to California and close to other states is pretty close leading up to 2004. After this, gun related homicide shoot up in cities close to Texas, New Mexico, and Arizona, while not changing much in cities close to California.

We will dig more deeply into this case/paper in week 13. But it seems reasonably convincing evidence that, at least in this context, increased access to assault weapons leads to more murders.

## Our goals more generally

Let's step back and think of the pluses and minuses of the approach from this paper compared to just looking at the scatter plots we started with. The main advantage is that we are closer to what we’ll call an “all else equal” comparison, where we can be reasonably confident attributing changes in relative homicide rates in parts of Mexico that had an influx of assault weapons vs those that didn’t to the influx of assault weapons. In other words, we can be pretty sure that this correlation does in fact tell us about what caused what. 

On the other hand, by restricting our analysis to the parts of Mexico that border the US at a particular time period, and studying an increase in a particular kind of weapon, we might not necessarily learn too much about the broader question of whether access to guns leads to more homicides in general. Maybe the causal effect is different for different kinds of guns, or in countries that aren’t in the middle of a major drug war. Overall, we learn different things from these different analyses, and the goal of this course is to help you understand how much we should learn from different kinds of evidence.

Why did we spend a good chunk of our first chapter going through this example? Not because this is a course about gun violence, or because I think it is a case where we totally nailed down the research question we care about with perfect precision. Rather, I think it highlights a lot of the key ideas we are going to explore this semester.

First, if we want to understand how the world works, we need to combine both theory and evidence. This is not just true when studying politics; it is the essence of the scientific method. We can come up with tons of theories that seem plausible if we are unconstrained by evidence, and just looking at data alone can never tell us what is driving the patterns we observe. 

Second, when theorizing and looking at data, we will be particularly interested in causal relationships. Science is mainly about asking “why” questions, and the answer to “why” questions often starts with “because”. That isn’t a coincidence. We also care about causal questions for practical purposes: if we want to know whether a policy is good or bad, we need to know what the causal effect is on outcomes we care about (homicides, economic growth, individual liberty). And any time we want to make an individual choice (whether to buy a gun, what to eat for lunch), we are implicitly asking about the causal effect of that choice on some outcome (our safety, how tasty our lunch is).

Third, doing good social science requires diverse skillsets. The majority of this course will be about analyzing data, and that is a key skill. But we also need to have people who are good at collecting data, designing surveys to measure political concepts, good at theorizing, and people who know a lot about the political contexts they study, for example to know that comparing groups of Mexican towns before and after 2004 might tell us something about the causal effect of making assault weapons more available. 

Science is almost always a collaborative process. One of the most important scientific discoveries of the recent past was the detection of the Higgs Boson, and the paper announcing this had 5,000 coauthors. We don’t typically go this far in political science--many of our largest conferences don't even have 5,000 participants--but teams of 2-5 are becoming the norm. This is because we all have different skills, and doing good work requires combining these. 

A lot of the material in this course will be about why social science is so hard. For example, if correlation isn’t causation, how can we learn about causation? But we’ll also try to convince you that when we combine good theory, evidence, and case-specific knowledge, we can learn useful things about how the political world works. And this course aims to start you on the path of being able to do this yourself, and with collaborators. 


## Caveats

A few caveats.

First, we are going to focus more on the quantitative side of political science. One reason is that this is the dominant mode of political science these days. If you go and pick up a top political science journal, it will be filled with statistical tests (and, to a lesser extent, mathematical theories). In order to be able to read contemporary research, you need to learn these methods.

It’s also the case that data analysis is becoming a bigger part of how people discuss politics outside of academia. I have certainly seen a major trend since I was students in your shoes, when it seemed more common to hear a sentiment that there is no way that data could be used to study something as messy as political behavior. This view is on it’s way out. Many newspapers now employ people who specialize in data analysis and visualization.

This is arguably even more true in companies and organizations where you may end up working. 

Without any training in statistics and other technical disciplines, all of this can seem impenetrable. But we can get you on a good trajectory to understanding these methods if you get through this course. If you are really interested in doing innovative work on your own, you’ll need to read more books and take more classes. But this should be a good start. Ultimately, we want you to be empowered to know how to collect and analyze data on your own, and to avoid being fooled by others using these techniques.

We are also going to be focused more on *positive* political science, or trying to understand the world as it is. This often contrasts with normative analysis, or how we think the world ought to be. Even if we knew for sure that more restrictive gun control policy would lead to fewer deaths, one could respond that there is a cost to having the government tell people what they are allowed to own. Positive analysis can never tell us how we should trade off deaths and freedom, though it may tell us roughly how many deaths would be saved by a particular infringement.

This line is often blurry: normative considerations will affect the kinds of things we want to study. Our opening question could reasonably be phrased as “should we pass more restrictive gun control policy”, which is really a blend of a normative and positive question. The positive can tell us what the effects of a policy can be, but we need the normative to tell us whether these effects are in aggregate good or bad.

Still, we should be careful to combine normative and positive analysis in a good way. A good combination is use normative concerns to tell us what positive analysis we should do. We might think that mass shootings are an important problem, and so we want to seek out positive analysis to inform how to reduce them. What we want to avoid is having our normative views distort the positive conclusion we reach. We don’t want to say “I think that gun control is bad, and so I want to find evidence to support that view.” (Or, “I think that gun control is good…”) Our values shouldn’t tell us what the data should say, but the data can say how we can pursue these values.
